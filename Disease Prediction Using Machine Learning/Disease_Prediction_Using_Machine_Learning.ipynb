```python
df = pd.read_csv("Training.csv")
```

```python
import seaborn as sns 
sns.heatmap(df.isnull())
```

```python
df.drop("Unnamed: 133", axis=1, inplace=True)
```

```python
listToStr = ', '.join([str(elem) for elem in df["prognosis"].unique()])
print("The Diseases lists are:", listToStr)
print(df["prognosis"].unique())
```

```python
X= df.drop("prognosis", axis=1)
Y = df["prognosis"]
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
X_train.shape
```

```python
type(X_train)
```

```python
df_testing = pd.read_csv("Testing.csv")
x_testing = df_testing.drop("prognosis", axis=1)
y_testing = df_testing["prognosis"]
```

<h3> Decision Tree </h3>

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
predictions_test = dt.predict(x_testing) 
```

```python
print(classification_report(y_testing, predictions_test))
```

```python
sns.heatmap(confusion_matrix(y_testing, predictions_test))
```

```python
print("Accuracy on Testing:",accuracy_score(y_testing, predictions_test))
```

```python
df_testing.join(pd.DataFrame(dt.predict(x_testing), columns=["predictions"]))[["prognosis","predictions"]]
```

<h3> Random Forest </h3>

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=500)
rf.fit(X_train, y_train)    
predictions = rf.predict(x_testing)
print("-------------Classifications Report-------------")
print(classification_report(y_testing, predictions))
print("-------------Confusion matrix-------------")
sns.heatmap(confusion_matrix(y_testing, predictions))
print("Accuracy on Testing:",accuracy_score(y_testing, predictions))
```

<h3>Neural Network </h3>


```python
y_train_dummy = pd.get_dummies(y_train)
```

```python
y_train_dummy
```

```python
y_train_dummy.shape[1]
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense
```

```python
model = Sequential()
model.add(Dense(64, activation="relu", input_dim=X_train.shape[1]))
model.add(Dense(48, activation="relu"))
model.add(Dense(y_train_dummy.shape[1], activation="softmax"))

model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
model.summary()
```

```python
history = model.fit(X_train, y_train_dummy, epochs=50, batch_size=30)
```

```python
print("ANN Accuracy:", history.history["accuracy"][-1])
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y_test_encoding = encoder.fit_transform(y_testing)
```

```python
print(y_test_encoding)
```

```python
#y_testing, predictions
predictions = model.predict(x_testing)
predictions = [np.argmax(i) for i in predictions]
print("Test Sets accuracy", accuracy_score(y_test_encoding, predictions))
sns.heatmap(confusion_matrix(y_test_encoding, predictions))
```
